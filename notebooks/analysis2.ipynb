```python
## Task 2

# load data
df = pd.read_csv('../data/raw/EMLSdata_10Aug.csv')

# rename columns to clearer names
df = df.rename(columns={'Lake_ID' : "LakeID", 'Altitude_m' : 'Altitude', 'MaximumDepth_m' : 'MaximumDepth', 'MeanDepth_m' : 'MeanDepth', 
                        'SurfaceTemperature_C' : 'SurfaceTemperature', 'Chlorophylla_ugL' : 'Chlorophylla', 'Chlorophyllb_ugL' : 'Chlorophyllb',
                        'Chlorophyllc2_ugL' : 'Chlorophyllc'})

# keep columns we want
df = df[['LakeID', 'Date', 'LakeName', 'LabName', 'Country', 'Latitude', 'Longitude', 'Altitude', 'MaximumDepth', 'MeanDepth', 'Chlorophylla', 'Chlorophyllb', 'Chlorophyllc']]

#drop na values
df = df.dropna(how="any")

# sort by Altitude and reset index
df = df.sort_values("Altitude").reset_index(drop=True)


# create date columns
df = df.assign(Year = lambda x : df['Date'].str[:4])
df = df.assign(Month = lambda x : df['Date'].str[5:7])
df.head()
```

```python
## Task 3
def load_and_process(pathname):

  # Method Chain 1 (Load data and deal with missing data)

  df1 = (
      pd.read_csv('../data/raw/EMLSdata_10Aug.csv')
      .rename(columns={'Lake_ID' : "LakeID", 'Altitude_m' : 'Altitude', 'MaximumDepth_m' : 'MaximumDepth', 'MeanDepth_m' : 'MeanDepth', 
                        'SurfaceTemperature_C' : 'SurfaceTemperature', 'Chlorophylla_ugL' : 'Chlorophylla', 'Chlorophyllb_ugL' : 'Chlorophyllb',
                        'Chlorophyllc2_ugL' : 'Chlorophyllc'})
      [['LakeID', 'Date', 'LakeName', 'LabName', 'Country', 'Latitude', 'Longitude', 'Altitude', 'MaximumDepth', 'MeanDepth', 'Chlorophylla', 'Chlorophyllb', 'Chlorophyllc']]
      .dropna(how="any")
      .sort_values("Altitude")
      .reset_index(drop=True)
  )

  # Method Chain 2 (Create new columns, drop others, and do processing)

  df2 = (
      df1
      .assign(Year = lambda x : df1['Date'].str[:4])
      .assign(Month = lambda x : df1['Date'].str[5:7])
  )

  return df2

load_and_process('../data/raw/EMLSdata_10Aug.csv')
```
