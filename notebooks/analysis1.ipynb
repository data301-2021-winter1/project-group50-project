## Part 2: Clean Data

```python
df_clean = df.drop(["Lake_ID","LabName","Altitude_m", "SamplingDepth_m", "ThermoclineDepth_m", "SecchiDepth_m"], axis=1)
df_clean
```

```python
# generate count of duplicate entries
if len(df_clean[df_clean.duplicated()]) > 0:
    print("No. of duplicated entries: ", len(df_clean[df_clean.duplicated()]))
    print(df_clean[df_clean.duplicated(keep=False)].sort_values(by=list(df_clean.columns)).head())
else:
    print("No duplicated entries found")
```

```python
# Find the sum of null values in each column 
df_clean.isnull().sum()
```

```python
# Drop null values
df_no_null = df_clean.dropna(axis=0)
df_no_null
```

```python
# Columns that contain more than 50% of null values will be dropped
df_new_version = df_no_null.drop(columns=df_no_null.columns[df_no_null.eq(0).mean()>0.5])
df_new_version
```

```python
# Reset index
df_new_version = df_new_version.reset_index()
df_new_version
```

```python
# Get rid of index column 
df_new_version = df_new_version.drop('index',axis =1)
df_new_version
```

## Part 3: Process Data

```python
# Rename columns 
df_new_version = df_new_version.rename(columns = {
               "MaximumDepth_m": "Maximum Lake Depth (m)",
               "MeanDepth_m": "Mean Depth Lake (m)",
               "EpilimneticTemperature_C":"Epilimnetic Temperature (C)",
               "TP_mgL":"Phosphate Concentration (mg/L)",
               "SurfaceTemperature_C": "Lake Surface Temperature (C)",
               "TN_mgL": "Nitrogen Concentration (mg/L)",
                "NO3NO2_mgL": "Concentration of Nitrates and Nitrites (mg/L)",
                "NH3_mgL": "Concentration of Ammonia (mg/L)",
                "PO4_ugL": "Concentration of Ortho-phosphate (ug/L)",
                "Chlorophylla_ugL": "Concentration of Chlorophyll-a (ug/L)",
                "Chlorophyllb_ugL" : "Concentration of Chlorophyll-b (ug/L)",
                "Zeaxanthin_ugL" : "Concentration of Zeaxanthin (ug/L)",
                "Alloxanthin_ugL" : "Concentration of Alloxanthin (ug/L)",
                "Chlorophyllc2_ugL" : "Concentration of Chlorophyll-c2 (ug/L)",
                "MC_YR_ugL" : "Concentration of Microcystin YR (ug/L)",
                "MC_dmLR_ugL" : "Concentration of Microcystin dmLR (ug/L)",
                "MC_LR_ugL" : "Concentration of Microcystin LR (ug/L)",
                "LakeName" : "Lake Name",
                         
              })    

df_new_version
```

```python
# Create a new column that gives the total concentration of photosynthetic pigments
df_new_version["Total Concentration Photosynthetic Pigments (ug/L)"] = 0.2*df_new_version["Concentration of Chlorophyll-a (ug/L)"] + 0.2* df_new_version["Concentration of Chlorophyll-b (ug/L)"] + 0.2* df_new_version["Concentration of Zeaxanthin (ug/L)"] + 0.2 * df_new_version["Concentration of Alloxanthin (ug/L)"]+ 0.2 * df_new_version["Concentration of Chlorophyll-c2 (ug/L)"]
df_new_version
```

```python
# Create a new column that gives the total concentration of cyanobacterial hepatotoxins
df_new_version["Total Concentration Cyanobacterial Hepatotoxins (ug/L)"] = 0.33*df_new_version["Concentration of Microcystin YR (ug/L)"] + 0.33* df_new_version["Concentration of Microcystin dmLR (ug/L)"] + 0.33* df_new_version["Concentration of Microcystin LR (ug/L)"] 
df_new_version
```

## Part 4: Wrangle Data

```python
#Re-organize columns to put total concentration of photosynthetic pigments and cyanobacterial hepatotoxins at the beginning. Also switch Lake Name and Country columns for clarity.
df_new_version = df_new_version[["Date", "Country", "Lake Name", "Total Concentration Cyanobacterial Hepatotoxins (ug/L)", "Total Concentration Photosynthetic Pigments (ug/L)", "Latitude", "Longitude", "Maximum Lake Depth (m)", "Mean Depth Lake (m)", "Lake Surface Temperature (C)", "Epilimnetic Temperature (C)", "Phosphate Concentration (mg/L)", "Nitrogen Concentration (mg/L)", "Concentration of Nitrates and Nitrites (mg/L)", "Concentration of Ammonia (mg/L)", "Concentration of Ortho-phosphate (ug/L)", "Concentration of Chlorophyll-a (ug/L)", "Concentration of Chlorophyll-b (ug/L)", "Concentration of Zeaxanthin (ug/L)", "Concentration of Alloxanthin (ug/L)", "Concentration of Chlorophyll-c2 (ug/L)", "Concentration of Microcystin YR (ug/L)", "Concentration of Microcystin dmLR (ug/L)", "Concentration of Microcystin LR (ug/L)"]]
df_new_version 
```

```python
# Sort rows by ascending order according to Total Concentration Cyanobacterial Hepatotoxins (ug/L)
df_new_version2 = df_new_version.sort_values("Total Concentration Cyanobacterial Hepatotoxins (ug/L)")
df_new_version2
```
